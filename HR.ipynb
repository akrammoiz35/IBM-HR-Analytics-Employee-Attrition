# Analysis & Data Visualization

```python
# Count the occurrences of each attrition value
attrition_count = df['Attrition'].value_counts()

# Print the attrition counts
attrition_count
```

```python
# Extract attrition categories and counts
attrition = attrition_count.index
counts = attrition_count.values

# Create a bar chart
fig, ax = plt.subplots()
bars = ax.bar(attrition, counts, color=["teal", "cyan"])  # Set bar colors

# Calculate and format percentages for bar labels
percentages = [(count / counts.sum() * 100) for count in counts]
labels = [f'{p:.0f}%' for p in percentages]

# Add percentage labels to bars with padding
ax.bar_label(bars, labels=labels, padding=3)

# Set chart title and labels
ax.set_title('Attrition')
ax.set_xlabel('Attrition Category')
ax.set_ylabel('Count')

# Display the plot
plt.show()
```

```python
# Count the occurrences of each business travel category
business_travel_counts = df['BusinessTravel'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(business_travel_counts,
        labels=business_travel_counts.index,  # Use category labels
        autopct='%0.2f%%',  # Format percentage with two decimals
        startangle=140,  # Start angle of the pie chart
        colors=["teal", "cyan", "turquoise"])  # Set colors for slices

# Set chart title
plt.title('Distribution of BusinessTravel Categories')

# Display the plot
plt.show()
```

```python
# Count employees by department and sort
total_department = df['Department'].value_counts().sort_index()

# Calculate average daily rate by department and sort
department_avg_rate = df.groupby('Department')['DailyRate'].mean().sort_index()

# Create a dual-axis subplot figure
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot department counts with bar chart
total_department.plot(kind='bar', color="teal", width=0.8, ax=axs[0])
axs[0].set_title('Departments')
axs[0].set_xlabel("\nDepartments names")  # Add newline for better spacing
axs[0].set_ylabel("Departments counts")
axs[0].set_xticks(range(len(total_department)))
axs[0].set_xticklabels(total_department.index, rotation=45)

# Plot average daily rate by department with bar chart
department_avg_rate.plot(kind='bar', color="cyan", width=0.8, ax=axs[1])
axs[1].set_title('Average Daily Rate by Department')
axs[1].set_xlabel('Department')
axs[1].set_ylabel('Average Daily Rate')
axs[1].set_xticks(range(len(department_avg_rate)))
axs[1].set_xticklabels(department_avg_rate.index, rotation=45)

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
Gender_counts = df['Gender'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(Gender_counts,
        labels=Gender_counts.index,  # Use category labels
        autopct='%0.2f%%',  # Format percentage with two decimals
        startangle=140,  # Start angle of the pie chart
        colors=["teal", "cyan", "lightcoral"])  # Set colors for slices

# Set chart title
plt.title('Distribution of Employees by Gender')

# Display the plot
plt.show()
```

```python
# Group the DataFrame by 'Gender' and count the occurrences of each 'Attrition' value
attrition_by_gender = df.groupby('Gender')['Attrition'].value_counts()

# Print the results
print(attrition_by_gender)
```

```python
# Group the DataFrame by 'Age' and count the occurrences of each 'Attrition' value
attrition_by_age = df.groupby('Age')['Attrition'].size().sort_values(ascending=False)[:10]

# Print the results
print(attrition_by_age)
```

```python
# Group the DataFrame by 'Gender' and count occurrences of each 'Attrition' value
attrition_by_gender = df.groupby('Gender')['Attrition'].value_counts()

# Group the DataFrame by 'Age' and count occurrences of each 'Attrition' value
attrition_by_age = df.groupby('Age')['Attrition'].size().sort_values(ascending=False)[:10]

# Create a dual-axis subplot figure
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot attrition by gender with bar chart
attrition_by_gender.plot(kind='bar', color='teal', width=0.8, ax=axs[0])
axs[0].set_title('Attrition by Gender')
axs[0].set_xlabel("Gender")
axs[0].set_ylabel("Attrition")
axs[0].set_xticks(range(len(attrition_by_gender)))
axs[0].set_xticklabels(attrition_by_gender.index, rotation=45)

# Plot attrition by age with bar chart
attrition_by_age.plot(kind='bar', color='cyan', width=0.8, ax=axs[1])
axs[1].set_title('Attrition by Age (Top 10)')
axs[1].set_xlabel('Age')
axs[1].set_ylabel('Attrition')
axs[1].set_xticks(range(len(attrition_by_age)))
axs[1].set_xticklabels(attrition_by_age.index, rotation=45)

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
